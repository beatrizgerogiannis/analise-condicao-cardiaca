IMPORTS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Análise Inicial

```python
df = pd.read_csv('UCMF_raw.csv')
pd.set_option('display.max_columns', None)

import pandas as pd

print("--- PRIMEIRAS 5 LINHAS ---")
display(df.head())

print('--- Colunas e tipos ---')
display(df.dtypes)

print('--- Valores únicos (top 10) por coluna ---')
for c in df.columns:
    print(f'\n-- {c} --')
    display(df[c].value_counts(dropna=False).head(10))

print('--- Resumo de valores ausentes ---')
display(df.isnull().sum().sort_values(ascending=False))

# Estatísticas numéricas e distribuição de idade
num = df.select_dtypes(include=['number'])
if not num.empty:
    print('--- Resumo numérico ---')
    display(num.describe())

    if 'IDADE' in df.columns:
        plt.figure(figsize=(8,4))
        sns.histplot(df['IDADE'].dropna(), kde=True)
        plt.title('Distribuição de IDADE')
        plt.show()

        plt.figure(figsize=(10,4))
        sns.boxplot(x=df['IDADE'].dropna())
        plt.title('Boxplot de IDADE')
        plt.show()

# Correlações entre numéricas
corr = num.corr()
if not corr.empty:
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Matriz de correlação (numéricas)')
    plt.show()

print('--- Amostra aleatória (10 linhas) ---')
display(df.sample(10))
```

## Tratamento Inicial


```python
import numpy as np

# 1. Padronização da Variável Alvo e Sexo
df['NORMAL X ANORMAL'] = df['NORMAL X ANORMAL'].str.title().replace('Normais', 'Normal')
df['SEXO'] = df['SEXO'].str.upper().str[0] # Padroniza pegando apenas a primeira letra

# 2. Remoção de Outliers Extremos (Substituindo por NaN)
# Filtrando idades para a faixa do estudo (0 a 19 anos)
df.loc[(df['IDADE'] < 0) | (df['IDADE'] > 19), 'IDADE'] = np.nan 

# Tratando Peso e Altura impossíveis
df.loc[df['Peso'] <= 0, 'Peso'] = np.nan
df.loc[df['Altura'] <= 0, 'Altura'] = np.nan

# Tratando Pressão Arterial irreal (ex: maior que 300)
df.loc[df['PA SISTOLICA'] > 300, 'PA SISTOLICA'] = np.nan
    
# 1. SEXO: Converter "Indeterminado" para NaN
df['SEXO'] = df['SEXO'].replace('I', np.nan)

# 2. SOPRO: Padronizar as categorias (corrigindo minúsculas)
df['SOPRO'] = df['SOPRO'].replace({
    'sistólico': 'Sistólico',
    'contínuo': 'Contínuo'
})

# 3. PPA: Remover erro de fórmula e transformar em NaN
df['PPA'] = df['PPA'].replace('#VALUE!', np.nan)

# 4. PULSOS: Padronizar as categorias (corrigindo maiúsculas)
df['PULSOS'] = df['PULSOS'].replace({
    'NORMAIS': 'Normais',
    'AMPLOS': 'Amplos'
})

# Seleciona apenas as colunas de texto (categóricas)
cat = df.select_dtypes(include=['object', 'category'])

if not cat.empty:
    print('--- Quantidade de valores únicos por coluna categórica ---')
    # .nunique() conta os valores únicos. O sort_values organiza do maior para o menor.
    display(cat.nunique().sort_values(ascending=False))

print('--- Detalhamento de valores únicos (Colunas Categóricas) ---')
for c in cat.columns:
    quantidade_unicos = df[c].nunique()
    print(f'\n-- {c} (Total de valores únicos: {quantidade_unicos}) --')
    # O dropna=False permite que você veja também quantos NaNs existem
    display(df[c].value_counts(dropna=False))
    
print('--- Resumo de valores ausentes ---')
display(df.isnull().sum().sort_values(ascending=False))

```
