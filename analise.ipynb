## Análise Inicial

```python
df = pd.read_csv('/content/UCMF_raw.csv')
pd.set_option('display.max_columns', None)

import pandas as pd

print("--- PRIMEIRAS 5 LINHAS ---")
display(df.head())

print("\n--- INFORMAÇÕES GERAIS ---")
df.info()

print("\n--- RESUMO ESTATÍSTICO (Numéricos) ---")
display(df.describe())

print("\n--- RESUMO ESTATÍSTICO (Textos/Categóricos) ---")
display(df.describe(include='object'))

print("\n--- VALORES AUSENTES POR COLUNA ---")
nulos = df.isnull().sum()
porcentagem = (df.isnull().mean() * 100).round(2) 

tabela_nulos = pd.DataFrame({
    'Total Ausentes': nulos, 
    'Porcentagem (%)': porcentagem
})

tabela_nulos = tabela_nulos[tabela_nulos['Total Ausentes'] > 0].sort_values(by='Porcentagem (%)', ascending=False)

if not tabela_nulos.empty:
    display(tabela_nulos) 
else:
    print("A base esta limpa.")

print("\n--- LINHAS DUPLICADAS ---")
print(f"Total de duplicadas: {df.duplicated().sum()}")

print("\n--- LINHAS COM SEXO 'Inderteminado' ---")

filtro_indeterminado = df[df['SEXO'] == 'Indeterminado']

print(f"Total de registros encontrados: {len(filtro_indeterminado)}\n")

display(filtro_indeterminado)
```
